<h1>Scraping the Wikipedia page for the table of postal codes of Canada</h1>
<h3> BeautifulSoup Library of Python is used for web scraping of table from the Wikipedia. The title of the webpage is printed to check if the page has been scraped successfully or not. Then the table of postal codes of Canada is printed.</h3>

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

<h1> The html table is converted to Pandas DataFrame for cleaning and preprocessing. </h1>

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

<h1> Data preprocessing and cleaning. </h1>

```python
# Dropping the rows where Borough is 'Not assigned'
df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2
```

```python
df2.shape
```

<h1> Importing the csv file conatining the latitudes and longitudes for various neighbourhoods in Canada </h1>

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
geo_data=pd.read_csv(lat_lon)
#geo_data.head(10)
```

<h1> Merging the two tables for getting the Latitudes and Longitudes for various neighbourhoods in Canada. </h1>

```python
#lat_lon.rename(columns={'Postal Code':'Postcode'},inplace=True)
df3 = pd.merge(df2,geo_data,on='Postal Code')
df3.head(20)
```


<h1> The notebook from here includes the Clustering and the plotting of the neighbourhoods of Canada which contain Toronto in their Borough. </h1>
<h1> Getting all the rows from the data frame which contains Toronto in their Borough. </h1>

```python
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

<h1> Visualizing all the Neighbourhoods of the above data frame using Folium </h1>

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

<h1> Using KMeans clustering for the clsutering of the neighbourhoods. </h1>

```python
k=5
toronto_clustering = df4.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
#df4.insert(0, 'Cluster Labels', kmeans.labels_)
df4
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
